# Processing NMR stacks with NMRpy

## Introduction

Below is an updated system to process NMR stacks using the NMRpy library. This has the advantage of increased performance and more intelligent peak-picking and integration over the MATLAB implementation.

Parameters to edit:

```python
dirname = "/Users/apavao/Desktop/NMR runs/20210103_13CGlc"
lb_factor = 5.0   # line-broadening
zf_factor = 3     # zero-filling
```

## Importing

```python
import glob
import os

import nmrglue
import nmrpy
```

## Loading the data

```python
fid_array = nmrpy.data_objects.FidArray.from_path(dirname)
```

## Processing the FID spectra

```python
fid_array.emhz_fids(lb=lb_factor)  # line-broadening
for i in range(zf_factor):         # zero-filling
    fid_array.zf_fids()
fid_array.ft_fids()                # Fourier transform
fid_array.fid00.phaser()         # Baseliner
fid_array.baseline_correct_fids()  # Baseline correct
fid_array.baseliner_fids()         # Baseliner
fid_array.baseline_correct_fids()  # Baseline correct
```
